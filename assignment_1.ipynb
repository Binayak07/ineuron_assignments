```python
#3>
"""
Three data types: String, Integer and Boolean.
String contains the string values, like name or place or anything inside quotes. (type is str)
Integer contains numberic values like 1, -9, 100 etc. And not floating point values like 16.02 etc (type is int)
Boolean contains truthy or falsy values, its value can be either True or False. (type is bool)
"""
print(type(100)) # int
print(type(True)) # bool
print(type("Ram")) #str
```

```python
#4>
"""
An expression is made up of variables and operators (+-=/* etc).
All expressions yeild some value.
"""
# Example:

val1 = 100
val2 = 50
name= "binayak"
title = "choudhary"

# expressions:::

quo = val1/val2 # this will give us quotient
sum = val1+val2 # this will give us sum
isGreater = val1>val2 # this will compare and return accordingly
full_name = name+title # this will concat two strings
print(quo,sum,isGreater, full_name)
```
